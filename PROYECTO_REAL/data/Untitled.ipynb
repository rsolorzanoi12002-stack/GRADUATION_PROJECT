```python
import sqlite3
import pandas as pd

db_path = "database/calidad_aire.db"
conn = sqlite3.connect(db_path)

# 1. Ver todas las tablas
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tablas = cursor.fetchall()
print("üìã TABLAS EN LA BASE DE DATOS:")
for tabla in tablas:
    print(f"  ‚Ä¢ {tabla[0]}")

print("\n" + "="*60)

# 2. Ver estructura de alertas_sistema (si existe)
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='alertas_sistema'")
if cursor.fetchone():
    cursor.execute("PRAGMA table_info(alertas_sistema)")
    columnas = cursor.fetchall()
    print("üìä ESTRUCTURA de alertas_sistema:")
    for col in columnas:
        print(f"  ‚Ä¢ {col[1]} ({col[2]})")
    
    # Ver algunos datos
    print("\nüìù DATOS en alertas_sistema:")
    df = pd.read_sql_query("SELECT id, nivel, procesada, fecha_procesada FROM alertas_sistema LIMIT 5", conn)
    print(df.to_string())
else:
    print("‚ùå alertas_sistema NO EXISTE")

print("\n" + "="*60)

# 3. Ver estructura de alertas (si existe)
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='alertas'")
if cursor.fetchone():
    cursor.execute("PRAGMA table_info(alertas)")
    columnas = cursor.fetchall()
    print("üìä ESTRUCTURA de alertas:")
    for col in columnas:
        print(f"  ‚Ä¢ {col[1]} ({col[2]})")
    
    # Ver algunos datos
    print("\nüìù DATOS en alertas:")
    try:
        df = pd.read_sql_query("SELECT * FROM alertas LIMIT 5", conn)
        print(df.to_string())
    except:
        print("  (Error leyendo datos - estructura corrupta)")
else:
    print("‚úÖ alertas NO EXISTE (o ya fue eliminada)")

conn.close()
```
